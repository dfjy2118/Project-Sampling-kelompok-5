## 3. Analisis Demografi – Distribusi Usia

```python
import pandas as pd
import plotly.express as px

# Membersihkan kolom usia
df['Usia_Clean'] = (
    df['Usia: __________ Tahun ']
    .astype(str)
    .str.replace(' tahun', '', regex=False)
    .str.strip()
    .astype(int)
)

# Membuat histogram interaktif
fig = px.histogram(
    df,
    x='Usia_Clean',
    nbins=6,
    marginal='box',  # tambahan boxplot di atas
    opacity=0.8,
    color_discrete_sequence=['skyblue'],
    title='Distribusi Usia Responden'
)

# Mempercantik tampilan
fig.update_layout(
    template='plotly_white',
    xaxis_title='Usia',
    yaxis_title='Jumlah Responden',
    title_x=0.5,
    font=dict(size=14),
    bargap=0.1,
    hovermode='x unified'
)

# Menampilkan grafik
fig.show()
```

## Interpretasi: Distribusi Usia Responden

Visualisasi histogram 'Distribusi Usia Responden' dengan tambahan boxplot marginal menunjukkan beberapa poin penting:

*   **Mayoritas Responden Dewasa Muda:** Sebaran data dengan jelas menunjukkan bahwa sebagian besar partisipan survei adalah individu dalam rentang usia remaja akhir hingga dewasa muda. Hal ini terbukti dari puncak distribusi di area tersebut dan konsentrasi data dalam boxplot.
*   **Implikasi Demografis:** Dominasi kelompok usia ini mengindikasikan bahwa studi ini cenderung mencerminkan kebiasaan dan persepsi bermain game online dari segmen demografi yang paling aktif dalam ekosistem game. Penemuan ini penting untuk mengontekstualisasikan hasil survei lainnya, karena pengalaman bermain game dan dampaknya mungkin berbeda di kelompok usia lain.
*   **Rentang Usia yang Terwakili:** Meskipun didominasi oleh dewasa muda, adanya rentang usia yang cukup luas (seperti yang ditunjukkan oleh jangkauan histogram dan boxplot) memberikan gambaran bahwa game online menarik minat dari berbagai kalangan, meskipun dengan konsentrasi yang berbeda-beda.

Secara keseluruhan, analisis distribusi usia ini memperkuat pemahaman bahwa game online adalah aktivitas yang sangat populer di kalangan dewasa muda, dan temuan survei ini akan paling relevan untuk kelompok demografi tersebut.

*   **4-6 jam:** Sebanyak **13 responden (24.53%)** melaporkan durasi tidur antara 4 hingga 6 jam. Kelompok ini mengalami kekurangan tidur jika dibandingkan dengan rekomendasi kesehatan.

## 7. Hubungan Durasi Bermain Game dengan Durasi Tidur

```python
cross_tab = pd.crosstab(
    df['Rata-rata durasi bermain game dalam satu hari: '],
    df['Rata-rata durasi tidur Anda dalam satu hari (24 jam): ']
)

fig = px.imshow(
    cross_tab,
    text_auto=True,
    aspect="auto",
    color_continuous_scale='Blues',
    title='Hubungan Durasi Bermain Game dan Durasi Tidur'
)

fig.update_layout(
    xaxis_title='Durasi Tidur',
    yaxis_title='Durasi Bermain Game',
    title_font_size=16
)
fig.show()
```

## Interpretasi:

Heatmap dan tabel persentase ini digunakan untuk melihat hubungan antara durasi bermain game dan durasi tidur responden. Semakin lama waktu bermain game, ada kecenderungan durasi tidur responden menjadi lebih sedikit. Berikut adalah beberapa observasi kunci:

*   **Durasi Bermain < 2 jam:**
    *   Mayoritas responden (sekitar **50-60%**) dalam kelompok ini masih berhasil mendapatkan tidur `6-8 jam` atau `>8 jam`. Hanya sebagian kecil yang tidur `<4 jam` atau `4-6 jam`.
    *   Ini menunjukkan bahwa durasi bermain game yang rendah umumnya tidak terlalu mengganggu pola tidur yang sehat.

*   **Durasi Bermain 2-4 jam:**
    *   Terjadi peningkatan proporsi responden yang tidur lebih sedikit. Misalnya, persentase yang tidur `4-6 jam` mungkin lebih tinggi dibandingkan kelompok `<2 jam`.
    *   Namun, masih banyak juga yang mendapatkan tidur cukup (`6-8 jam`).

*   **Durasi Bermain 4-6 jam:**
    *   Di kelompok ini, kita akan melihat pergeseran yang lebih jelas menuju durasi tidur yang lebih pendek. Persentase responden yang tidur `6-8 jam` atau `>8 jam` kemungkinan menurun, sementara yang tidur `4-6 jam` atau bahkan `<4 jam` bisa meningkat signifikan.

*   **Durasi Bermain > 6 jam:**
    *   Kelompok dengan durasi bermain game tertinggi ini kemungkinan besar akan menunjukkan persentase tertinggi responden yang tidur `4-6 jam` atau `<4 jam`. Ini adalah indikator kuat bahwa bermain game secara berlebihan secara langsung berkorelasi dengan kurangnya waktu tidur.

Secara keseluruhan, analisis ini mengkonfirmasi hipotesis bahwa ada korelasi negatif antara durasi bermain game dan durasi tidur; semakin lama seseorang bermain game, semakin besar kemungkinan mereka memiliki durasi tidur yang lebih pendek.

## 11. Analisis Korelasi Pertanyaan Skala Likert

```python
likert_columns = [
    'Saya merasa sulit untuk berhenti bermain game meskipun tubuh sudah terasa sangat lelah',
    'Saya sering menambah durasi bermain karena merasa waktu yang ada selalu kurang',
    'Saya merasa gelisah, cemas, atau mudah marah jika tidak bisa mengakses game dalam sehari',
    'Saya tetap bermain game meskipun tahu ada tugas, pekerjaan, atau tanggung jawab yang terbengkalai',
    'Bermain game adalah cara utama saya untuk melarikan diri dari tekanan, stres, atau rasa bosan',
    'Saya sering tidur di atas jam 01.00 dini hari karena keasyikan bermain game.',
    'Saya sering menunda waktu tidur hanya untuk menyelesaikan satu sesi/pertandingan game lagi.',
    'Saya merasa durasi tidur harian saya berkurang drastis sejak sering bermain game online.',
    'Saya sering merasa mengantuk atau sulit berkonsentrasi saat beraktivitas di siang hari akibat begadang main game.',
    'Saya merasa kualitas tidur saya buruk (sering terbangun atau tidak segar saat bangun) karena kebiasaan bermain game sebelum tidur.'
]

# Mengubah tipe data menjadi numerik
for col in likert_columns:
    df[col] = pd.to_numeric(df[col], errors='coerce')

# Menghitung matriks korelasi
corr_matrix = df[likert_columns].corr()

# Singkat nama kolom untuk label heatmap agar tidak terlalu panjang/tumpang tindih
short_labels = [f"Q{i+1}" for i in range(len(likert_columns))]

fig = px.imshow(
    corr_matrix,
    x=short_labels,
    y=short_labels,
    text_auto=".2f", # Menampilkan angka 2 desimal di dalam kotak
    color_continuous_scale='RdBu_r', # Warna merah (positif) ke biru (negatif)
    title='Korelasi Antar Pertanyaan Skala Likert (Q1 - Q10)'
)

fig.update_layout(title_font_size=16)
fig.show()
```

## Interpretasi:

Heatmap korelasi ini menampilkan hubungan linear antar sepuluh pertanyaan Skala Likert (Q1-Q10) yang berkaitan dengan kebiasaan bermain game dan dampaknya terhadap tidur serta konsentrasi.

*   **Skala Warna:**
    *   Warna merah (mendekati 1) menunjukkan korelasi positif yang kuat. Artinya, jika responden cenderung setuju pada satu pertanyaan, mereka juga cenderung setuju pada pertanyaan lain yang berkorelasi positif kuat.
    *   Warna biru (mendekati -1) menunjukkan korelasi negatif yang kuat. Artinya, jika responden cenderung setuju pada satu pertanyaan, mereka cenderung tidak setuju pada pertanyaan lain yang berkorelasi negatif kuat.
    *   Warna mendekati putih (mendekati 0) menunjukkan korelasi yang lemah atau tidak ada.

*   **Diagonal Utama:** Selalu bernilai 1.00 karena ini adalah korelasi pertanyaan dengan dirinya sendiri.

Dengan menganalisis nilai-nilai korelasi ini, kita dapat mengidentifikasi pertanyaan-pertanyaan mana yang memiliki hubungan paling erat. Misalnya, korelasi positif yang tinggi antara pertanyaan tentang "sulit berhenti bermain game" (Q1) dan "sering tidur di atas jam 01.00 dini hari" (Q6) dapat mengindikasikan bahwa kesulitan menghentikan game sangat berhubungan dengan kebiasaan begadang. Hal yang sama berlaku untuk dampak seperti sulit konsentrasi (Q9) dan kualitas tidur buruk (Q10) yang kemungkinan besar berkorelasi positif dengan pertanyaan-pertanyaan yang mengukur kebiasaan bermain game berlebihan.

(Catatan: Label kolom dipersingkat menjadi Q1-Q10 agar tampilan grafik interaktif tetap rapi dan tidak terpotong).

## 11b. Alternatif Analisis Hubungan: Box Plot Interaktif

```python
# Kita ambil salah satu dampak: Dampak terhadap Konsentrasi (Q9)
# Dan kita bandingkan berdasarkan kelompok Durasi Bermain Game

fig = px.box(
    df,
    x='Rata-rata durasi bermain game dalam satu hari: ',
    y='Saya sering merasa mengantuk atau sulit berkonsentrasi saat beraktivitas di siang hari akibat begadang main game.',
    title='Sebaran Dampak Konsentrasi Berdasarkan Durasi Bermain Game',
    color='Rata-rata durasi bermain game dalam satu hari: ',
    color_discrete_sequence=px.colors.qualitative.Safe
)

fig.update_layout(
    xaxis_title='Durasi Bermain Game per Hari',
    yaxis_title='Skala Mengantuk / Sulit Konsentrasi (1-5)',
    showlegend=False,
    title_font_size=16
)
fig.show()
```

## Interpretasi:

Box plot ini menggambarkan hubungan antara rata-rata durasi bermain game per hari dengan skala perasaan mengantuk atau kesulitan berkonsentrasi saat beraktivitas di siang hari. Setiap 'kotak' (box) pada plot menunjukkan sebaran jawaban (interquartile range) untuk kelompok durasi bermain game tertentu, dengan garis tengah menandakan median.

Dari visualisasi ini, Anda dapat melihat:
*   **Tren Median:** Bagaimana nilai tengah (median) dari 'Skala Mengantuk / Sulit Konsentrasi' berubah seiring dengan peningkatan 'Durasi Bermain Game'. Umumnya, kita akan mencari apakah median cenderung meningkat pada durasi bermain game yang lebih lama.
*   **Sebaran Data:** Seberapa luas kotak (IQR) dan jangkauan 'whisker' untuk setiap durasi bermain. Ini menunjukkan variasi dampak konsentrasi dalam setiap kelompok. Jika kotak dan whisker lebih panjang pada durasi game yang lebih lama, ini berarti dampak negatifnya lebih bervariasi atau lebih ekstrem.
*   **Outlier:** Titik-titik di luar 'whisker' menunjukkan outlier, yaitu responden yang melaporkan dampak konsentrasi yang jauh berbeda dari mayoritas dalam kelompok durasi bermain game mereka.

Secara keseluruhan, plot ini membantu mengidentifikasi apakah durasi bermain game yang lebih panjang cenderung dikaitkan dengan dampak negatif yang lebih besar (skala yang lebih tinggi) pada konsentrasi dan tingkat kantuk responden.

Secara keseluruhan, meskipun ada yang tidak bermain game atau tidak begadang, sebagian besar responden menunjukkan bahwa daya tarik game yang kuat (kesenangan, tantangan, sosial), kebutuhan akan relaksasi, dan ketersediaan waktu di malam hari adalah faktor utama yang membuat mereka terus bermain hingga larut malam.
